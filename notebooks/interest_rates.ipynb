# US Interest Rates

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import datetime
```

```python
fred = pd.read_csv('../price_ranges/interest_rates.csv')
fred
```

```python
index = fred.iloc[:, 0]
```

```python
fred.index = index
```

```python
fred.drop('observation_date', axis=1, inplace=True)

fred.head()
```

```python
fred.plot(kind='line',
          xlabel='Date',
          ylabel='Federal Funds Rate (%)',
          title='Federal Funds Rate from 2020 - 2025')
plt.grid(alpha=.4)
plt.show()
```
